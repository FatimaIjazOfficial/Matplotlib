# Matplotlib Data Visualization With NumPY

## IMPORT LIBRARIES

```python
import matplotlib.pyplot as plt
import numpy as np
```

## 📊 Histogram: Group Comparison

We generate two normally distributed datasets:
- Group A: mean = 70, std = 10
- Group B: mean = 80, std = 12

Then we compare their distributions using histograms.

```python
group_a = np.random.normal(70, 10, 1000)
group_b = np.random.normal(80, 12, 1000)
```

```python
plt.figure(figsize=(7,5))

plt.hist(group_a, bins=30, density=True, alpha=0.6, label="Group A")
plt.hist(group_b, bins=30, density=True, alpha=0.6, label="Group B")

plt.xlabel("Score")
plt.ylabel("Density")
plt.title("Distribution Comparison")
plt.legend()
plt.savefig("histogram_comparison.png", dpi=300)
plt.show()
```

## 📊 Subplots Overview

We create multiple visualizations in one figure:
- Line Plot
- Scatter Plot
- Histogram
- Bar Chart

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(8, 6))

# Line Plot
axs[0, 0].plot(x, y)
axs[0, 0].set_title("Line Plot")

# Scatter Plot
axs[0, 1].scatter(x, y)
axs[0, 1].set_title("Scatter Plot")

# Histogram
axs[1, 0].hist(np.random.randn(2000), bins=20)
axs[1, 0].set_title("Histogram")

# Bar Chart
categories = ["A", "B", "C"]
values = [5, 3, 7]
axs[1, 1].bar(categories, values)
axs[1, 1].set_title("Bar Chart")

plt.savefig("subplot_grid.png", dpi=300)

plt.tight_layout()
plt.show()
```

## 📈 Customized Sine Plot

We plot a sine wave with:
- Custom ticks
- Labels
- Legend
- Saved as PNG file

```python
x = np.linspace(0, 10, 50)
y = np.sin(x)
```

```python
fig, ax = plt.subplots(figsize=(6, 4))

ax.plot(x, y, label="sin(x)", linewidth=2)
ax.set_xticks(range(0, 11, 2))

plt.xticks(rotation=30)

ax.set_title("Customized Sin Wave Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.legend(loc="upper right")

plt.tight_layout()
plt.savefig("sine_plot.png", dpi=300)
plt.show()
```
