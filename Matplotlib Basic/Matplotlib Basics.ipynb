# 📊 Matplotlib Basics
This notebook includes:
1. Blank figure
2. Line plot (Sales trend)
3. Scatter plot (Temperature vs Ice cream sales)
4. Bar chart (Monthly sales comparison)
5. Histogram (Age distribution)
6. Combined 2x2 subplot with all graphs together

```python
import matplotlib.pyplot as plt
```

## Blank Figure

```python
fig, ax = plt.subplots()
ax.set_title("Blank Figure Example")
ax.set_xlabel("X-axis Label")
ax.set_ylabel("Y-axis Label")
fig.savefig("blank_plot.png", dpi=100)
plt.show()
```

## Line Plot (Sales Trend) 

```python
days = [1, 2, 3, 4, 5, 6, 7]
sales = [120, 150, 180, 170, 200, 220, 210]
```

```python
plt.figure()
plt.plot(days, sales, marker='o', linewidth=1, alpha=0.8, label="Daily Sales")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.title("Sales Trend over a Week")
plt.legend()
plt.savefig("Line Plot (Sales Trend).png", dpi=100)
plt.show()
```

## Scatter Plot

```python
temperature = [20, 22, 25, 27, 30, 32, 35]
icecream_sales = [50, 65, 80, 90, 120, 140, 160]
```

```python
plt.figure()
plt.scatter(
    temperature,
    icecream_sales,
    marker='x',
    alpha=0.7,
    label="Sales vs Temperature"
)
plt.xlabel("Temperature (°C)")
plt.ylabel("Ice Cream Sales")
plt.title("Relation Between Temperature and Ice Cream Sales")
plt.legend()
plt.savefig("Scatter Plot.png", dpi=100)
plt.show()
```

## Bar Chart
Bar chart compares categories. Here we compare monthly sales.

```python
months = ["Jan", "Feb", "Mar", "Apr", "May"]
monthly_sales = [200, 250, 300, 280, 350]
```

```python
plt.figure()
plt.bar(months, monthly_sales, color='green', alpha=0.7, label="Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly Sales Comparison")
plt.legend()
plt.savefig("Bar Chart.png", dpi=100)
plt.show()
```

## Histogram

```python
ages = [18, 19, 20, 21, 22, 23, 24, 25, 22, 21, 20, 19, 18, 23, 24, 25, 26, 27, 28, 22]
```

```python
plt.figure()
plt.hist(ages, bins=6, color='blue', alpha=0.7, edgecolor='black', label="Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Histogram of Ages")
plt.legend()
plt.savefig("Histogram.png", dpi=100)
plt.show()
```

## Combined 2x2 Subplots (All Graphs Together)

- Line plot
- Scatter plot
- Bar chart
- Histogram


```python
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# Line Plot
axes[0, 0].plot(days, sales, marker='o', color="orange")
axes[0, 0].set_title("Line Plot")

# Scatter Plot
axes[0, 1].scatter(temperature, icecream_sales,color="blue")
axes[0, 1].set_title("Scatter Plot")

# Bar Chart
axes[1, 0].bar(months, monthly_sales,color="red")
axes[1, 0].set_title("Bar Chart")

# Histogram
axes[1, 1].hist(ages, bins=6,color="green")
axes[1, 1].set_title("Histogram")

plt.savefig("All Graphs.png", dpi=100)
plt.tight_layout()
plt.show()
```
